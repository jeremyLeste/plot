```python
pdd6e=np.loadtxt('/home/crct/CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/2/6e/validation/6epddgate.csv', delimiter=",")
pdd9e=np.loadtxt('/home/crct/CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/2/9e/validation/9epddgate.csv', delimiter=",")
pdd12e=np.loadtxt('/home/crct/CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/2/12e/validation/12epddgate.csv', delimiter=",")
pdd18e=np.loadtxt('/home/crct/CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/2/18e/validation/18epddgate.csv', delimiter=",")
```

```python
pdd6 = Scatter(x = pdd6e[:,0],
               y = pdd6e[:,1],
               name ='6 MeV',
               mode = 'lines')
pdd9 = Scatter(x = pdd9e[:,0],
                 y = pdd9e[:,1],
                 name ='9 MeV',
                 mode = 'lines')

pdd12= Scatter(x = pdd12e[:,0],
                 y = pdd12e[:,1],
                 name ='12 MeV',
                 mode = 'line')
pdd18e[:,1]=pdd18e[:,1]/(pdd18e[:,1].max())
pdd18= Scatter(x = pdd18e[:,0],
                 y = pdd18e[:,1],
                 name ='18 MeV',
                 mode = 'line')



layout = go.Layout(
    xaxis=dict(
    range=[0,130])    
)

fig = dict(data=[pdd6,pdd9, pdd12, pdd18], layout=layout)
iplot(fig, image='svg')
```
