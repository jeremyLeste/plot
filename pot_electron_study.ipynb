```python
import sys , os 
import numpy as np
import math as m
import pandas as pd

import plotly.offline as py

import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
```

```python
energy='18'
heter='b'
depth1='20'
depth2='40'

name1=energy+heter+depth1
name2=energy+heter+depth2

gate1=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gate_'+name1+'.csv', delimiter=",")
mes1=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/mes_'+name1+'.csv', delimiter=",")
emc1=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/emc_'+name1+'.csv', delimiter=",")
#gammaG1=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gammaGate_'+name1+'.csv', delimiter=",")
#gammaE1=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gammaEmc_'+name1+'.csv', delimiter=",")

gate2=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gate_'+name2+'.csv', delimiter=",")
mes2=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/mes_'+name2+'.csv', delimiter=",")
emc2=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/emc_'+name2+'.csv', delimiter=",")
#gammaG2=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gammaGate_'+name2+'.csv', delimiter=",")
#gammaE2=np.loadtxt('/media/sf_CRCT/JupyterNotebook/data/plot/gamaIndex/elecArticle/curie/final/'+energy+'e/gammaEmc_'+name2+'.csv', delimiter=",")
```

```python
mes1X=mes1[:, 0]
mes1Y=mes1[:, 1]
mes2X=mes2[:, 0]
mes2Y=mes2[:, 1]

gate1X2=gate1[:, 0]
gate1Y2=gate1[:, 1]
gate2X2=gate2[:, 0]
gate2Y2=gate2[:, 1]

emc1X2=emc1[:, 0]
emc1Y2=emc1[:, 1]
emc2X2=emc2[:, 0]
emc2Y2=emc2[:, 1]

gate2Y2=gate2Y2*0.5
emc2Y2=emc2Y2*0.5
mes2Y=mes2Y*0.5
```

```python
# Plot profil or PDD with plotly      
trace1 = go.Scatter(
    x = mes1X,
    y = mes1Y,
    mode='line',
    name='measure '+depth1+'mm',
    line = dict(
        color = ('rgba(10, 20, 22, 10)'),
        width = 1.5,
        shape = 'hv'
    )
)

trace2 = go.Scatter(
    x = mes2X,
    y = mes2Y,
    mode='line',
    name='measure '+depth2+'mm',
    line = dict(
        color = ('rgba(10, 20, 22, 10)'),
        width = 1.5,
        dash = 'dot',
        shape = 'hv'
    ),
    showlegend = True
)


trace3 = go.Scatter(
    x = gate1X2,
    y = gate1Y2,
    mode='markers',
    name='GATE '+depth1+'mm',
    line = dict(
        color = ('rgba(0, 92, 245, 100)'),
        width = 0.5,
        shape = 'hv'
         
    ),
    marker = dict(
   symbol = 'square-open-dot',
   ),
)

trace4 = go.Scatter(
    x = gate2X2,
    y = gate2Y2,
    mode='markers',
    name='GATE '+depth2+'mm',
    line = dict(
        color = ('rgba(0, 92, 245, 100)'),
        width = 0.5,
        shape='hv'
         
    ),
     marker = dict(
   symbol = 'dot',
   ),
   
    showlegend = True
)

trace5 = go.Scatter(
    x = emc1X2,
    y = emc1Y2,
    mode='markers',
    name='eMC '+depth1+'mm',
    line = dict(
        color = ('rgba(214, 39, 40, 100)'),
        width = 0.5,
        shape='hv'
         
    ),
    marker = dict(
   symbol = 'square-open-dot',
)
)


trace6 = go.Scatter(
    x = emc2X2,
    y = emc2Y2,
    mode='markers',
    name='eMC '+depth2+'mm',
    line = dict(
        color = ('rgba(214, 39, 40, 100)'),
        width = 0.5,
        shape='hv'
         
    ),
    marker = dict(
   symbol = 'dot',
),
    showlegend = True
)




layout1= go.Layout(
    title= 'Dose Profile - '+energy+' MeV',
    titlefont=dict(size=22),
    xaxis=dict(
        range=[-100, 100],
        title='Off axis distance (mm)',
        titlefont=dict(
            size=16
        )
),      
    yaxis=dict(
        title='Relative Dose',
        titlefont=dict(
        size=16)
    ),
    
     annotations=[
        dict(
            x=-25,
            y=1.08,
            xref='gateX2',
            yref='gateY2',
            showarrow=False, 
            text= depth1 +'mm',
            font=dict(
            size=15,
            )
     ),
        dict(
            x=-25,
            y=0.58,
            xref='gateX2',
            yref='gateY2',
            showarrow=False, 
            text= depth2 +'mm',
            font=dict(
            size=15,
            )
     ),
        dict(
            x=-25,
            y=emc1Y2.max()+0.2,
            xref='gateX2',
            yref='gateY2',
            showarrow=False, 
            text= 'Water',
            font=dict(
            size=20,
            )
        ),
        dict(
            x=25,
            y=emc1Y2.max()+0.2,
            xref='gateX2',
            yref='gateY2',
            showarrow=False, 
            text= 'Bone',
            font=dict(
            size=20,
            )
        ),
         
         
    ],
    
    
    showlegend = True
    
)
data1 = [trace1, trace2, trace3, trace4, trace5, trace6]
fig1= go.Figure(data=data1, layout=layout1)
py.plot(fig1, filename='plot_'+name1+name2, image='svg', auto_open=True)
```
